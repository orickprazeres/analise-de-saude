```python
def lastest_by_country(data):
    return data.groupby('Country/Region').sum()[last_date]

lastest_by_country(confirmed).head()
```

```python
def lastest_by_country(data):
    return data.groupby('Country/Region').sum().iloc[:,-1]

lastest_by_country(confirmed).head()
```

```python
display(lastest_by_country(confirmed).head())
display(lastest_by_country(deaths).head())
display(lastest_by_country(recovered).head())
```

```python
informatioin = [lastest_by_country(confirmed), lastest_by_country(deaths), lastest_by_country(recovered)]
combined = pd.concat(informatioin, axis = 1)
combined.columns = ('confirmed', 'deaths', 'recovered')
combined.head()
```

```python
plt.figure(figsize= (15,5))
combined.loc['Brazil'].plot(kind = 'bar')
plt.title('Resumo dos casos de COVID-19 no Brasil', fontsize = 16)
```

```python
def rename_location(location):
    if location == 'United States of America': 
        return 'US'
    if location == 'Republic of Korea': 
        return 'Korea, South'
    if location == 'Russian Federation': 
        return 'Russia'
    if location == 'TÃ¼rkiye':
        return 'Turkey'
    if location == 'Iran (Islamic Republic of)':
        return 'Iran'
    if location == 'Viet Nam':
        return 'Vietnam'
    return location


population_total_in_2022['location_for_who']= population_total_in_2022.index.map(rename_location)
population_total_in_2022.query("location_for_who == 'US'")
population_total_in_2022 = population_total_in_2022.set_index('location_for_who')
population_total_in_2022
```

```python
grouped_per_day = pd.DataFrame([confirmed.sum()[3:-2], recovered.sum()[3:-1], deaths.sum()[3:-1]],
            index = ['confimed', 'recovered', 'deaths']).T

grouped_per_day.head()
```

```python
grouped_per_day['letality_rate_1'] = grouped_per_day['deaths'] / grouped_per_day['confimed'] *100
grouped_per_day['letality_rate_2'] = grouped_per_day['deaths'] / (grouped_per_day['deaths'] + grouped_per_day['recovered']) *100

grouped_per_day.head()
```

```python
fig = plt.figure(figsize = (10,5))

grouped_per_day.letality_rate_1.plot()
ax = grouped_per_day.letality_rate_2.plot()

plt.title("Letalidade 'pessimista e otimista' no mundo todo, convergindo para o mesmo valor", fontsize = 14)
plt.legend(['Mortes / Casos confirmados', 'Mortes / (Mortes + Recuperados)'])
plt.grid(linestyle = '-')
ax.set_xlabel('Fonte: Curso www.alura.com.br e dados https://github.com/CSSEGISandData/COVID-19 ')
ax.set_ylabel('% letalidade')
plt.show()
```
